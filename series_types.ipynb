```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
%matplotlib inline
```

# Yearly (Cycle)

The Canadian lynx data set used in time series modelling is the annual record of the number of the Canadian lynx "trapped" in the Mackenzie River district of the North-West Canada for the period 1821â€“1934.

```python
ts_lynx = pd.Series.from_csv("data/canadian_lynx.csv")
```

```python
ts_lynx.head()
```

```python
ts_lynx[:50].plot(figsize=(15, 5))
```

```python
ts_lynx_ma = ts_lynx.rolling(3).mean()
ts_lynx_ma.plot(figsize=(15, 5))
```

```python
result = seasonal_decompose(ts_lynx)
result.plot()
```

# Weakly (Periodical)

Weekly data on US finished motor gasoline products supplied (in thousands of barrels per day) from February 1991 to May 2005.

```python
ts_gasoline = pd.Series.from_csv("data/gasoline.csv")
```

```python
print(len(ts_gasoline))
```

```python
ts_gasoline.plot(figsize=(15, 5))
```

```python
ts_gasoline_ma = ts_gasoline.rolling(5).mean()
ts_gasoline_ma.plot(figsize=(15, 5))
```

```python
result = seasonal_decompose(ts_gasoline)
result.plot()
```

```python
print("Input series:")
print(np.ptp(ts_gasoline))
print(np.max(ts_gasoline))
print(np.min(ts_gasoline))
print("Trend component:")
print(np.ptp(result.trend))
print(np.max(result.trend))
print(np.min(result.trend))
result.trend.plot(figsize=(15,5))
```
